```python
df0= pd.DataFrame(df.groupby('striker')['runs_off_bat'].sum()).reset_index()
```

```python
df1= pd.DataFrame(df.groupby('striker')['ball'].count()).reset_index()
```

```python
df2= df0.merge(df1,on = 'striker', how='left')
df2= df2[df2.runs_off_bat >= 500]
df2= df2[df2.ball <= 500]
df2.head()
```

```python
len(df2)
```

```python
df2['strike_rate']= df2['runs_off_bat']/df2['ball']*100
df2
```

```python
df2.reset_index(inplace=True, drop=True)
df2.head(18)
```

```python
df2=df2.sort_values("runs_off_bat")
df2
```

```python
plt.title('Batsmen with more than 500 runs scored from IPL 2020-2022', fontsize=20 )
plt.barh(df2["striker"],df2["runs_off_bat"])

plt.xlabel('Runs scored')
plt.show()
```

```python
df2=df2.sort_values("strike_rate")
df2
```

```python
plt.title('Batsmen strike rate with more than 500 runs scored from IPL 2020-2022', fontsize=20 )
plt.barh(df2["striker"],df2["strike_rate"])
plt.xlabel('strike_rate')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(18,9))
plt.scatter(df2.strike_rate, df2.runs_off_bat)

for i in range(len(df2)):
    if df2["striker"][i]=="MS Dhoni":
        plt.text(df2['strike_rate'][i]+1 , df2['runs_off_bat'][i]-1 , df2['striker'][i],color= 'maroon')
    else:
        plt.text(df2['strike_rate'][i] + 1, df2['runs_off_bat'][i] - 1, df2['striker'][i] )
        
plt.title('Batsmen with more than 500 runs scored with less than 500 balls(Finishers) from IPL 2020-2022', fontsize=26 )
plt.xlabel('Strike rate', fontsize=16)
plt.ylabel('Runs scored', fontsize=16)
plt.axvline(130, ls='--', color='black')
plt.axhline(500, ls='--', color='black')
plt.show()
```

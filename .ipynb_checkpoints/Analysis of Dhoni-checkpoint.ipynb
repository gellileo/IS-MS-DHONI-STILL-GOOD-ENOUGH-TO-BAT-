```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)  
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', -1)
```

```python
df=pd.read_csv("IPL_ball_by_ball_updated.csv")
```

```python
df.head()
```

```python
df = df[df['wides'].isnull()]
df.tail()
```

```python
df= df[(df.season>=2020)&(df.season<=2022)]
df.tail()
```

```python
mdf=df[df['striker']== 'MS Dhoni']
mdf.head()
```

```python
len(mdf)
```

```python
sum(mdf['runs_off_bat'])
```

```python
(sum(mdf['runs_off_bat'])/len(mdf))*100
```

```python
sum(df['player_dismissed'] == 'MS Dhoni')
```

```python
df0= pd.DataFrame(df.groupby('striker')['runs_off_bat'].sum()).reset_index()
```

```python
df1= pd.DataFrame(df.groupby('striker')['ball'].count()).reset_index()
```

```python
df2= df0.merge(df1,on = 'striker', how='left')
df2= df2[df2.runs_off_bat >= 500]
df2= df2[df2.ball <= 500]
df2.head()
```

```python
len(df2)
```

```python
df2['strike_rate']= df2['runs_off_bat']/df2['ball']*100
```

```python
df2.reset_index(inplace=True, drop=True)
df2.head(18)
```

# visualization

```python
df2.groupby(['striker'])['runs_off_bat'].sum().sort_values(ascending = False).plot(kind ='barh')
plt.xlabel('Runs scored')
plt.show()
```

```python
df2.groupby(['striker'])['strike_rate'].sum().sort_values(ascending = False).plot(kind ='barh')
plt.xlabel('strike_rate')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(18,9))
plt.scatter(df2.strike_rate, df2.runs_off_bat)

for i in range(len(df2)):
     plt.text(df2['strike_rate'][i] + 1, df2['runs_off_bat'][i] - 1, df2['striker'][i] )
        
plt.title('Batsmen with more than 500 runs scored with less than 500 balls(Finishers) from IPL 2020-2022', fontsize=26 )
plt.xlabel('Strike rate', fontsize=16)
plt.ylabel('Runs scored', fontsize=16)
plt.axvline(130, ls='--', color='black')
plt.axhline(500, ls='--', color='black')
plt.show()
```
